# Importing CSV into MySQL using Workbench:

MySQL Workbench is a versatile application that provides various functionalities for modeling, generating, and administering databases. The Community Edition of MySQL Workbench is freely available for download on the Oracle website.

CSV imports with MySQL Workbench, is super simple. Since CSV’s are probably the most common data interchange format, it goes without saying that importing CSV data into MySQL is a staple task for all DBA’s and Developers.

<a href="https://ibb.co/bL1XSDt"><img src="https://i.ibb.co/tcBD7FR/workbench-148737-1280.png" alt="workbench-148737-1280" border="0"></a>

This tutorial showcases two distinct methods to import data using MySQL Workbench. If you prefer a graphical user interface (GUI) approach, the first step involves establishing a remote connection to the MySQL database.

## Table Data Import Wizard.

To begin, in the navigation pane, perform a right-click on the table where you wish to import CSV data. Afterward, select the option **"Table Data Import Wizard."**

<a href="https://ibb.co/cyzmqqv"><img src="https://i.ibb.co/sgLB77P/img-1.png" alt="img-1" border="0"></a>

Select the Table Data Import Wizard, and you will be directed to the `Select File to Import` option or screen. From there, you have the choice to either manually input the file path or utilize the Browse button to locate and select the import file.

<a href="https://ibb.co/BwFkNrX"><img src="https://i.ibb.co/qxV65Wt/img-2.png" alt="img-2" border="0"></a>

To import the CSV file, click on the Browse button and select the desired file.

After selecting the file to import, the system presents a "Select Destination" screen with several options. On this screen, you can indicate the specific table you wish to use. There is also an option to create a new table by selecting the "Create New Table" checkbox. If the settings are to your liking, proceed by clicking the "Next" button.

<a href="https://ibb.co/GcFyBkQ"><img src="https://i.ibb.co/MM9rWpk/img-3.png" alt="img-3" border="0"></a>

MySQL Workbench Select Destination Table for Import

In this step, you are choosing the destination table for the import process.

After selecting the import settings, a `Configure Import Settings` screen will appear, allowing you to map the source column data (CSV header row) to the destination columns listed in the "Dest Column" section. For this process, I maintained all the default settings as shown in the screenshot below.

<a href="https://ibb.co/fDMY4CH"><img src="https://i.ibb.co/5KTWB5L/img-4.png" alt="img-4" border="0"></a>

In the user interface, you will find an "Import Data" tab. By clicking the "Next" button on this tab, the process of importing data begins. As each step is automatically executed in the background by the system, the corresponding radio select options, such as "Prepare Import" and "Import data file," are sequentially marked off.

Once all the above steps are completed, the "Cancel" button becomes inactive and cannot be clicked. At this point, your options are limited to either clicking the "<Back" or "Next>" button, depending on the step you wish to proceed with. To execute the actual import, you should click the "Next>" button.

<a href="https://ibb.co/YhfQsWn"><img src="https://i.ibb.co/kXqBCyb/img-5.png" alt="img-5" border="0"></a>

After the import process is finished, a confirmation message will be displayed on the Import Results screen, providing the following information:

- The successful upload of the file and the duration it took to complete the upload.
- The specific table that was utilized for the import operation.
- The total number of records (rows) that were successfully imported into the table.

Once the import is completed, you will see the message "Import complete" indicating that the process has finished successfully.
